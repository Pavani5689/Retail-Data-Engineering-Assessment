Reason:
This ensured accurate revenue calculations and maintained consistency with the organization's master product data.

```python
print("Missing Prices Before:")
print(retail_df["price"].isnull().sum())
```

Category Standardization

Solution:

Reason:
This ensured accurate aggregation and reporting in Power BI by preventing duplicate business categories.

Date Standardization

Problem:
transaction_date was stored as object datatype.

Solution:
The column was converted to datetime using Pandas to_datetime() function.

Reason:
This enables time-based analysis, trend reporting, and Power BI date intelligence features.

```python
print(retail_df["transaction_date"].dtype)

retail_df["transaction_date"] = pd.to_datetime(
    retail_df["transaction_date"],
    errors="coerce"
)

print(retail_df["transaction_date"].dtype)

print(retail_df["transaction_date"].isnull().sum())
```

```python
retail_df[["email","phone"]].head()
```

```python
def mask_email(email):
    if pd.isnull(email):
        return email

    name, domain = email.split("@")
    return name[:2] + "***@" + domain
```

```python
retail_df["email"] = retail_df["email"].apply(mask_email)
```

```python
def mask_phone(phone):
    phone = str(phone)

    return phone[:2] + "******" + phone[-2:]
```

```python
retail_df["phone"] = retail_df["phone"].apply(mask_phone)
```

```python
retail_df[["email","phone"]].head()
```
